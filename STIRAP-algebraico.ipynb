Se considera el sistema de 3 niveles en $\Lambda$ ( el orden de energías es: $\left| 1 \right\rangle,  \left| 2 \right\rangle,  \left| 3 \right\rangle)$. Los láseres tienen una desintonía y frecuencia de Rabi de $\Delta_p$ y $\Omega_p$ para el primer láser que va de $\left| 1 \right\rangle$ a $\left| 3 \right\rangle$ y $\Delta_s$, $\Omega_s$ para $\left| 2 \right\rangle$ a $\left| 3 \right\rangle$.


Su ecuación maestra es:

$$\dot{\rho}=-\frac{i}{\hbar}[H, \rho] + L(\rho),$$ 

$L(\rho)$ es el término de Lindblad. En este caso sólo se están considerando dos decaimientos: $$ \left| 3 \right\rangle \rightarrow \left| 1 \right\rangle, $$ y  $$\left| 3 \right\rangle \rightarrow \left| 2 \right\rangle .$$ 

Así el término de Lindblad se vuelve: $$L(\rho)=\gamma_{31}\big(\sigma_{13}\rho\sigma_{13}^\dagger-\frac{1}{2}\{\sigma_{13}^\dagger\sigma_{13},~\rho\}\big)+\gamma_{32}\big(\sigma_{23}\rho\sigma_{23}^\dagger-\frac{1}{2}\{\sigma_{23}^\dagger\sigma_{23},~\rho\}\big).$$ 

Las $\sigma$ son los operadores de aniquilación y creación así que: $$ \sigma_{13}=\left| 1 \right\rangle \left\langle 3 \right|,~ \sigma_{13}^\dagger=\left| 3 \right\rangle \left\langle 1 \right|,$$ $$\sigma_{23}=\left| 2 \right\rangle \left\langle 3 \right|,~ \sigma_{23}^\dagger=\left| 3 \right\rangle \left\langle 3 \right|.$$

Escribiendo esto de manera matricial

$$\sigma_{13} =  = \begin{bmatrix} 0 &0 &1 \\0&0&0\\ 0& 0& 0 \end{bmatrix},~\sigma_{23} = \begin{bmatrix} 0 &0 &0 \\0&0&1\\ 0&0& 0 \end{bmatrix},$$ y $$\sigma_{13}^\dagger \sigma_{13}=\sigma_{23}^\dagger \sigma_{23}=\begin{bmatrix} 0&0&0\\0&0&0\\0&0&1 \end{bmatrix},$$

entonces $$L_{31} = \gamma_{31}\big(\sigma_{13}\rho\sigma_{13}^\dagger-\frac{1}{2}\{\sigma_{13}^\dagger\sigma_{13},~\rho\}\big) = \begin{bmatrix} \gamma_{31} \rho_{13}&0&-\frac{1}{2}\gamma_{31}\rho_{13}\\0&0&-\frac{1}{2}\gamma_{31}\rho_{23}\\-\frac{1}{2}\gamma_{32}\rho_{31}&-\frac{1}{2}\gamma_{32}\rho_{32}&-\gamma_{31}\rho_{33} \end{bmatrix},$$

$$L_{32}=\gamma_{32}\big(\sigma_{23}\rho\sigma_{23}^\dagger-\frac{1}{2}\{\sigma_{23}^\dagger\sigma_{23},~\rho\}\big)=\begin{bmatrix} 0&0&-\frac{1}{2}\gamma_{32}\rho_{13}\\0&\gamma_{32}\rho_{33}&-\frac{1}{2}\gamma_{32}\rho_{23}\\-\frac{1}{2}\gamma_{32}\rho_{13}&-\frac{1}{2}\gamma_{32}\rho_{32}&-\gamma_{32}\rho_{33} \end{bmatrix},$$



El término de Lindblad queda como $$L(\rho)=\begin{bmatrix} \gamma_{31}\rho_{33} & 0 & -\frac{1}{2}(\gamma_{31}+\gamma_{32})\rho_{13} \\
0 & \gamma_{32}\rho_{33} & -\frac{1}{2}(\gamma_{31}+\gamma_{32})\rho_{23} \\
-\frac{1}{2}(\gamma_{31}+\gamma_{32})\rho_{31}  & -\frac{1}{2}(\gamma_{31}+\gamma_{32})\rho_{32}  & -\frac{1}{2}(\gamma_{31}+\gamma_{32})\rho_{33} 
\end{bmatrix},$$


Al resolver la ecuación maestra quedan 9 ecuaciones diferenciales con términos complejos. Como la matriz $\rho$ es hermitiana entonces las coherencias cumplen que $\rho_{ij}=\rho_{ji}^*$, de modo que que se reducen a 6 ecuaciones diferenciales complejas. Si se toman por separado los términos reales e imaginarios de las coherencias entonces terminan quedando 9 ecuaciones diferenciales reales. Las variables serán: $$\rho_{11},~\rho_{22},~\rho_{33},~Re\{\rho_{12}\},~Im\{\rho_{12}\},~Re\{\rho_{13}\},~Im\{\rho_{13}\},~Re\{\rho_{23}\}~\textrm{y}~Im\{\rho_{23}\},$$ y sus respectivas derivadas temporales.

Por la condición de adiabaticidad lo que haré será ir lentamente variando la intensidad de los pulsos, tan lento que cada vez que vario las frecuencias de Rabi el sistema llega a un estado estacionario en el cual $\dot{\rho}=0$

Sea P el vector columna con las $\rho$, $\dot{P}$ el vector columna de las $\dot{\rho}$, F la matriz con los coeficientes y C el vector constante. Entonces la ecuación matricial de arriba se puede abreviar como:

$$ \dot{P}=F \cdot P - C,$$

En el caso estacionario $\dot{P}=0$, entonces se puede despejar y encontrar que P queda:

$$P=F^{-1}\cdot{C}$$

Para lograr la transferencia adiabatica de población es necesario que los láseres se prendan y apaguen con un orden determinado; primero el láser $\Omega_s$ que acopla a ls niveles $\left| 2 \right\rangle$ y $\left| 3 \right\rangle$ y mientras se va apagando este láser y prendiendo el láser $\Omega_p$ que acopla $\left| 1 \right\rangle$ y $\left| 3 \right\rangle$. 

Ambos pulsos se modelaran con gaussianas y a continuación se muestran. Primero para el pulso $\Omega_p$:

```julia
function Ω_p(t)
    Tmax=1
    T=sqrt(2)/10*Tmax
    Ω=1
    Ω*exp(-((t-Tmax/2)-T)^2/T^2)
end

function Ω_p(t::Array) 
    R=[]
    for k=1:length(t)
        R=push!(R, Ω_p(t[k]))
    end
    R
end
```

Luego el pulso $\Omega_s$:

```julia
function Ω_s(t)
    Tmax=1
    T=(sqrt(2)/10)*Tmax
    Ω=1
    Ω*exp(-((t-Tmax/2)+T)^2/T^2)
end

function Ω_s(t::Array) 
    R=[]
    for k=1:length(t)
        R=push!(R,Ω_s(t[k]))
    end
    R
end
```

Y se grafican ambos pulsos:

```julia
Tmax=1
t=collect(linspace(0.0, Tmax, 1000))
y=Ω_p(t)
plot(t, Ω_p(t))
plot(t, Ω_s(t))
```

A continuación se ponen la perte del código que resulve el sistema de ecuaciones. La función A genera la matriz con los coeficientes y la función C genera al vector columna C.

```julia
function A(t)
    γ31=100
    γ32=100
    Ωp=1*Ω_p(t)
    Ωs=1*Ω_s(t)
    Δp=100
    Δs=100
    F=zeros(8,8)
    F[1,2], F[1,8] = γ32, 2*Ωs
    F[2,2], F[2,7], F[2,8] = -γ31-γ32, -2Ωp, 2Ωs
    F[3,6], F[3,7], F[3,8] = Δp-Δs, Ωs, Ωp
    F[4,4], F[4,6], F[4,7] = -0.5*γ31, Ωs, Ωp
    F[5,5], F[5,6], F[5,8] = -0.5*γ31, -Ωp, Δs
    F[6,3], F[6,4], F[6,5] = -Δp + Δs, -Ωs, Ωp
    F[7,1], F[7,2], F[7,3], F[7,4], F[7,7] = Ωp, 2Ωp, -Ωs, Δp, -0.5*γ31
    F[8,1], F[8,2], F[8,3], F[8,5], F[8,8] = -Ωs, Ωs, -Ωp, -Δs, -0.5*γ32
    F
end

function C(t)
    m=zeros(8,1)
    m[7,1]=Ω_p(t)
    m
end
    
```

Luego se saca la inversa de F y se multiplica por el vector C. Notese que aunque se toma el caso adiabatico tanto F como C son funciones del tiempo y varían con el prendido y pagado de los pulsos.

```julia
function P(C::Array,F::Array)
    inv(F)*C
end
    
```

```julia
Res=[]
Tmax=1
for k=1:length(t)
    Res=push!(Res,P(C(t[k]),A(t[k])))
end
Res;
```

El arreglo Res contiene la evolución temporal de la solución del tiempo 0 al 1. Extrayendo los coeficientes $\rho_22$ y $\rho_33$ a cada tiempo y graficándolos se puede observar como se mueven las poblaciones entre los niveles atómicos.

```julia
E2=[]
for i=1:length(t)
    E2=push!(E2,Res[i][1])
end
E2

E3=[]
for i=1:length(t)
    E3=push!(E3,Res[i][2])
end
E3;
```

En la siguiente grafica se muestra la transferencia adiabatica de poblaciones. En rojo se pone la poblacion del nivel 1, en verde la del nivel 2 y en azul la del nivel 1./

```julia
# Población en el nivel 3 (azul)
plot(t,E3)

# Población en el nivel 2 (verde)
plot(t,E2)

# Población en l nivel 1 (rojo)
plot(t,1-E2-E3)
```

Como la línea azul de la población del nivel 3 apenas se ve, a continuación se grafica para ver que efectivamente es cero siempre, nunca se pasa por este nivel.
